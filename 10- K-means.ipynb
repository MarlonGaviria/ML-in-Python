# k-medias

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
datos = make_blobs(n_samples= 200, n_features=2,centers=4)
```

```python
plt.scatter(datos[0][:,0],datos[0][:,1])
```

```python
km = KMeans(n_clusters=4)
```

```python
km.fit(datos[0])
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
datos[1]
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize = (12,4))
ax1.scatter(datos[0][:,0],datos[0][:,1], c=km.labels_)
ax1.set_title('algoritmo de K-means')

ax2.scatter(datos[0][:,0],datos[0][:,1], c=datos[1])
ax1.set_title('Datos originales')
```
