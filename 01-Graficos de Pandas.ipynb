# Graficos de Pandas

```python
import pandas as pd
import numpy as np
```

```python
dt = pd.DataFrame(np.random.randint(200, size = (50,4)), columns= ['a','b','c','d'])
```

```python
dt.head()
```

```python
dt['a'].hist()
```

```python
dt.b.hist(bins=30)
```

```python
dt.a.plot.hist()
```

```python
dt.plot.area(alpha = 0.5)
```

```python
dt.plot.bar(stacked = True)
```

```python
dt.plot.scatter(x='a',y='b',c='c', cmap = 'coolwarm')
```

```python
dt.plot.box()
```

```python
dt.plot.hexbin('a','b', gridsize=12)
```

```python
dt.plot.kde()
```

```python
dt.plot.density()
```
