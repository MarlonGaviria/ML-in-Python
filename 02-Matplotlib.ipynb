# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
y = x**2 
```

```python
plt.plot(x,y,'lightgreen')
plt.title('mi grafico')
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(y,x,'blue')
```

#### con programacion oriendata a objetos

```python
f = plt.figure()
g = f.add_axes([0.1,0.1,0.8,0.8])
g.plot(x,y,'r')
```

```python
f = plt.figure()
g = f.add_axes([0.1,0.1,0.9,0.9])
g.plot(x,y,'r')

g2 = f.add_axes([0.15,0.5,0.4,0.4])
g2.plot(x,y,'y')
```

## Multigraficos

```python
figura, grafico = plt.subplots(2,2)

grafico[0][0].plot(x,y,'g')
grafico[0][1].plot(x,y,'r')
grafico[1][0].plot(x,y,'b')
grafico[1][1].plot(x,y,'pink')

grafico[0][0].set_title('green')
grafico[0][1].set_title('red')
grafico[1][0].set_title('blue')
grafico[1][1].set_title('pink')
```

# Tamao del grafico

```python
fig = plt.figure(figsize=(8,2))
graf = fig.add_axes([0.1,0.1,0.9,0.9])
graf.plot(x,y,'r')
```

```python
figura = plt.figure()
g = figura.add_axes([0.1,0.1,0.9,0.9])
g.plot(x,x**2, label = 'cudrado')
g.plot(x,x**3, label = 'cubo')
g.legend(loc = 0)
```

## Tipo de linea (grosor transparencia y tipo)

```python
figura = plt.figure()
g = figura.add_axes([0,0,1,1])
g.plot(x,y,'red',linewidth=5, alpha = 0.3, linestyle = ':')
```

## Marcadores

```python
figura = plt.figure()
g = figura.add_axes([0,0,1,1])
g.plot(x,y,'green', marker = 'o', markersize=5)
```
