### Graficos de cuadricula

```python
flores = sns.load_dataset('iris')
flores
```

```python
sns.pairplot(flores, hue = 'species')
```

```python
g = sns.PairGrid(flores)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(flores, hue='species')
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(prop, col = 'time', row = 'smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(prop, col = 'time', row = 'smoker')
g.map(plt.scatter,'total_bill','tip')
```

# Graficos de Regresion

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = prop)
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = prop, hue='sex', markers=['o','x'])
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = prop, hue='sex', markers=['o','x'], scatter_kws={'s':80})
```

```python
sns.set_context('notebook',font_scale=1.5)
sns.lmplot(x = 'total_bill',y = 'tip',data = prop, col = 'day', row='sex')
```

```python
sns.set_context('notebook',font_scale=1.5)
sns.lmplot(x = 'total_bill',y = 'tip',data = prop, col = 'day',row = 'smoker', hue='sex',
          aspect= 0.8, size = 8)
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = prop, aspect= 0.8, size = 10)
plt.show()
```

## Estilos y colores

```python
sns.set_context('notebook',font_scale=1.5) # cambiar la escala de la letra
plt.figure(figsize = (10,4))            # tamano del grafico
sns.set_style('white')                  # estilo
sns.countplot(x='sex', data = prop)     
sns.despine(left = True,bottom = True)  # quitar el recuadro
```

```python
sns.countplot(x='sex', data = prop, palette=['red','yellow'],saturation=0.4) 
```
